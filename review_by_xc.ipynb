```python
from curl_cffi import requests
from bs4 import BeautifulSoup
import re
import time
import gen_css
import add_current
```

```python
rest = requests.get('https://direct.mit.edu/rest',impersonate = 'chrome110')
rest
```

```python
rest_data = BeautifulSoup(rest.text, 'html.parser')
rest_data
```

```python
current_link = rest_data.find(class_="view-current-issue")
current_link
```

```python
current_url = 'https://direct.mit.edu' + current_link.find('a').get('href')
current_url
```

```python
issue = re.search('iss.+',current_url)[0]
issue = re.sub(r"/", "_", issue)
issue
```

```python
current_data = requests.get(current_url)
current_data
```

```python
data = BeautifulSoup(current_data.text, 'html.parser')
data
```

```python
articles = data.find_all(class_='customLink item-title')
articles
```

```python
article_link = 'https://direct.mit.edu' + articles[0].find('a').get('href')
article_link
```

```python
article_requests = requests.get(article_link)
article_requests
```

```python
article_data = BeautifulSoup(article_requests.text, 'html.parser')
article_data 
```

```python
ori_title = article_data.find(class_='wi-article-title article-title-main').text
```

```python
title = re.sub(r'[\r\n]+', '', ori_title).strip()
title
```

```python
ori_name = article_data.find_all(class_='info-card-name')
ori_name
```

```python
# 遍历每个 div 元素，将文本内容添加到字符串中
names = ""
for div in ori_name:
    names += div.get_text(strip=True) + ", "
names[:-2]
```

```python
ori_abstract = article_data.find(class_='abstract')#.find('p').get_text()
print(ori_abstract)
```

```python
title = []
name = []
abstracts = []
for art_link in articles:
    try:
        art_link = 'https://direct.mit.edu' + art_link.find('a').get('href')
        article_requests = requests.get(art_link)
        article_data = BeautifulSoup(article_requests.text, 'html.parser')
        ori_title = article_data.find(class_='wi-article-title article-title-main').text
        title.append(re.sub(r'[\r\n]+', '', ori_title).strip())
        ori_name = article_data.find_all(class_='info-card-name')
        names = ""
        for div in ori_name:
            names += div.get_text(strip=True) + ", "
            names = names[:-2]
        name.append(names)
        abstract = article_data.find(class_='abstract')
        abstracts.append(abstract.find('p').get_text())
    except:
        pass
article = [*zip(title, name, abstracts)]
article
```

```python
with open('rest/REST_' + issue + '.html','w',encoding='UTF-8') as f:
    f.write(gen_css.gen_content_from_list(title))
    f.write(gen_css.gen_abstract_list(list(article)))
```

```python
add_current.add('REST_' + issue, time.ctime(), 'rest/REST_' + issue + '.html')
```
